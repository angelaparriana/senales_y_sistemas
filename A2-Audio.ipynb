### Audio en Python

En este notebook veremos como generar e importar datos desde un archivo de audio y como graficarlos y escucharlos.

En primer lugar generaremos un vector de valores aleatorios de un segundo de duración, a una tasa de muestreo de 44.100 Hz y lo escucharemos vía audio en Python.


```python
from IPython.display import Audio
import numpy as np

signal = np.random.random(44100)
Audio(signal, rate=44100)
```

Para importar un archivo de audio podemos hacer lo siguiente:


```python
%matplotlib inline
import matplotlib.pyplot as plt

from scipy.io import wavfile
fs, data = wavfile.read('./sonidos/Primera.wav')

plt.plot(data)
plt.show()
```

Y con el siguiente código podemos escuchar el sonido importado.

```python
Audio(data, rate=fs)
```
