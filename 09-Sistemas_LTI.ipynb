## El Teorema del núcleo de Schwartz

El teorema del núcleo de Schwartz, o  [Schwartz kernel theorem](https://en.wikipedia.org/wiki/Schwartz_kernel_theorem), si bien es un teorema bastante complejo y es uno de los resultados fundamentales en la teoría de Schwartz, en el contexto de este curso nos interesa en la siguiente formulación:

Para cada sistema lineal $L$ exise un único núcleo o kernel $h(t,\xi)$ tal que

$$ L\{f(t)\} = \int_{-\infty}^{\infty} h(t,\xi) f(\xi) d\xi $$

Es decir, esto significa que un sistema lineal siempre puede entenderse como la integral de la función a la cual se le aplica el sistema multiplicada por un kernel de dos variables. 


## Aplicación de un sistema lineal e invariante a una distribución

Supongamos que un sistema lineal $L$ se aplica sobre una distribución $T$ convolucionada con una función $f$. Gracias al teorema del núcleo de Schwartz podemos escribir:

$$ \langle L \{\mathbf{f \ast T}\}, \varphi \rangle = \langle \mathbf{f \ast T}, L\{\varphi\} \rangle  = \langle \mathbf{T}, \partial_{-1} f \ast L\{\varphi\} \rangle \hspace{2cm}\forall \varphi \in S$$

pero

$$ f \ast L\{\varphi\}(t) = \int_{-\infty}^{\infty} f(\tau) L\{\varphi\}(t-\tau) d\tau $$


y dado que $L$ es invariante:

$$ f \ast L\{\varphi\}(t) = \int_{-\infty}^{\infty} f(\tau) L\{\varphi(t-\tau)\} d\tau $$

y lineal:

$$  = L\{\int_{-\infty}^{\infty} f(\tau) \varphi(t-\tau) d\tau\} = L\{(f \ast \varphi) (t)\} $$

Por lo tanto, tenemos que 

$$ \langle L \{\mathbf{f \ast \mathbf{T}}\}, \varphi \rangle = \langle \mathbf{T}, \partial_{-1} L\{f \ast \varphi\} \rangle = \langle L\{\mathbf{T}\}, \partial_{-1} f \ast \varphi \rangle = \langle L\{\mathbf{T}\} \ast f, \varphi \rangle$$

Concluímos que:

$$ \boxed{L\{f \ast \mathbf{T}\} = L\{\mathbf{T}\} \ast f} $$

## Respuesta al impulso

El resultado anterior nos permite escribir uno de los principios más relevantes de la teoría de sistemas LTI. 

Tomando el caso anterior cuando $\mathbf{T} = \delta$, tenemos que

$$ \boxed{L\{f \ast \mathbf{\delta}\} = L\{\mathbf{\delta}\} \ast f} $$

Definimos $\mathbf{h} = L\{\mathbf{\delta}\}$, como aquella distribución que resulta de aplicar una distribución $\delta$ como entrada a un sistema LTI. Esta distribución se conoce como *respuesta al impulso*.

Ahroa, considerando que

$$ f = f \ast \delta$$

tenemos que la respuesta de un sistema LTI a una entrada $f$ cualquiera es:

$$ \boxed{L\{f\} = L\{f \ast \delta\} = f \ast L\{\delta\} = f \ast h} $$

¿Qué sacamos como conclusión?
<div class="alert alert-block alert-info">
<font color="red"><b>La respuesta de un sistema LTI a una entrada arbitraria $f$ se puede obtener mediante la convolución de $f$ con la respuesta al impulso $h$. Es importante entender que este resultado ha sido determinado considerando distribucioens y no funciones.</b></font> 
</div>

Sin embargo, en la práctica, podemos entender $h$ como una función, y podemos escribir sin problemas:

$$ g(t) = L\{f(t)\} = (h \ast f)(t) = \int_{-\infty}^{\infty} h(t-\xi)f(\xi)d\xi$$

### El caso de sistemas variantes

Si un sistema es lineal, pero variante, no es posible utilizar la expresión anterior. Lo que sucede en este caso es que la respuesta la impulso no es única, sino que depende del desplazamiento que experimente la señal. Pero si se puede utilizar la *integral de superposición*:

$$ g(t) = L\{f(t)\} = \int_{-\infty}^{\infty} h(t,\xi)f(t)dt $$

donde podemos ver que la respuesta al impulso $h(t,\xi)$ no sólo es una función de la variable independiente $t$ sino que también del deplazamiento $\xi$. 

<div class="alert alert-block alert-info">
Como podemos comprobar, este caso lo mismo que plantea el teorema del núcleo de Schwartz descrito anteriormente. El núcleo o kernel es simplemente la respuesta al impulso del sistema lineal.
</div>

## Sistemas descritos por ecuaciones diferenciales de coeficientes constantes

Si un sistema se puede describir mediante una ecuación diferencial de orden $n$, con coeficientes constantes, entonces se trata de un sistema LTI. 

### Relación con la convolución

Consideremos un sistema descrito por una ecuación diferencial de orden $n$ de la forma:

$$ \biggr[c_0 + c_1\frac{d}{dt} + \dots + c_n \frac{d^n}{dt^n} \biggr] g(t) = f(t) $$

donde $f(t)$ y $g(t)$ son funciones. 

El sistema puede ser escrito en término de convoluciones con derivadas de orden hasta $n$ del impulso, de la siguiente manera:

$$ \biggr[ c_0 \delta + c_1 \delta' + \dots + c_n \delta^{(n)} \biggr] \ast g(t) = f(t) $$

Este resultado implica que muchos sistemas físicos puede ser planteados como un problema de convolución, mediante una ecuación de la forma

$$ H \ast G = F $$

donde $H$ y $F$ son distribuciones y la incógnita (o salida) $G$ también es una distribución. La idea es usar distribuciones para poder utilizar conceptos como el impulso en la soluciones de estas ecuaciones. 

El problema se reduce a encontrar una solución y determinar si esta solución es única. Entonces, el problema se trata de encontrar 

$$ H^{\ast -1} $$ 

distribución que se conoce como la inversa de convolución de $H$ o bien la [función de Green](https://en.wikipedia.org/wiki/Green%27s_function) de $H$, tal que

$$ \boxed{H \ast H^{\ast -1} =  H^{\ast -1} \ast H = \delta} $$ 

<div class="alert alert-block alert-info">
En este contexto, la función de Green, es equivalente con la idea de respuesta al impulso. Sin embargo, la función de Green típicamente se asocia a la solución de ecuaciones diferenciales, mientras que el concepto de respuesta al impulso se asocia a sistemas LTI. No todos los sistemas LTI son descritos por ecuaciones diferenciales.
</div>

### Ejemplo:

Consideremos la ecuación diferencial de primer orden

$$ f'(t) + a f(t) = 0 $$

la cual queremos resolver para el caso $t>0$ y con la condición inicial $f(0) = f_0$. 

Veamos como se puede resolver esta ecuación en el contexto de las distribuciones. Primero, necesitamos extender una distribucion a partir de la función $f$, la cual denotaremos $T_f$. 

Entonces, necesitamos determinar cual es la ecuación que debe satisfacer $T_f$. Como lo que queremos obtener es una solución para $t > 0$, con un valor inicial de $f_0$, necesariamente $T_f$ debe tener una discontinuidad, es decir debe estar multiplicada por un escalón. Entonces su derivada es 

$$ (T_f)' = T_{f'} + f_0 \delta = -a T_f + f_0 \delta $$

Es decir, tenemos la siguiente ecuación diferencial para la distribución $T_f$:

$$ (T_f)' + a T_f = f_0 \delta $$

Esta ecuación se puede escribir como una convolución de la forma

$$ (\delta' + a\delta) \ast T_f = f_0\delta $$

La distribución de Green de esta ecuación puede probarse que es 

$$ G = e^{-a t} u $$

donde $u$ es la distribución escalón, cuya derivada es:

$$ G' = -ae^{-at} + \delta $$ 

Para comprobar que es una solución podemos convolucionar la función de Green con el núcleo $\delta' + a\delta$:

$$ G = e^{-a t} u \ast \delta' + a\delta = -ae^{-at} + \delta + a e^{-at} = \delta $$

Para calcular la salida en el caso particular del ejemplo, solo necesitamos convolucionar con la respuesta al impulso, que en este caso de la función de Green:

$$ T_f = G \ast f_0 \delta = f_0 e^{-at} u $$

## Solución de ecuaciones diferenciales mediante distribuciones
Consideremos un sistema lineal de entrada $f$ y salida $g$ relacionados por la ecuación diferencial

\begin{equation}
  L g = f  
\end{equation}

donde $L$ es el operador diferencial de la forma:

$$  L = \sum_{i \in I} C_i \frac{d^{\alpha_i}}{dt^{\alpha_i}} $$


donde $\alpha_i \in \mathbb{N}$,  $I$ es de cardinalidad finita y $C_i \in \mathbb{C}$. Esta ecuación se puede extender a una forma más generalizada utilizando distribuciones temperadas:

\begin{equation}
  L \mathbf{g} = f   \;\;\;\;\;  L = \sum_{i \in I} C_i D^{\alpha_i}
\end{equation}

La distribución temperada $\mathbf{h} \in \mathcal{S}'(\mathbb{R},\mathbb{C})$ es una solución fundamental del operador diferencial $L$ si 

$$ L(\mathbf{h})= \mathbf{\delta} \;\;\; \in   \mathcal{S}'(\mathbb{R},\mathbb{C})$$

Esto implica que para cada $f \in \mathcal{S}(\mathbb{R},\mathbb{C}) $, la solución de la ecuación diferencial $L \mathbf{u} = f$ está dada mediante la convolución:

$$ \mathbf{g} = f \ast \mathbf{h} $$ 

En efecto, se tiene que: 
$$ L (\mathbf{g}) = \sum_{i \in I} C_i D^{\alpha_i} (f \ast \mathbf{h}) =  f \ast \sum_{i \in I} C_i D^{\alpha_i} \mathbf{h} = f \ast L(\mathbf{h})  $$ 

y dado que $ L(\mathbf{h}) = \delta $:
 
entonces:

$$ \boxed{L (\mathbf{g}) = f \ast \delta = f} $$

## Propiedades de los sistemas LTI

Tal como estipulamos anteriormente, un sistema LTI se caracteriza por su respuesta al impulso, que denotaremos por $h(\cdot)$. En las siguientes propiedades $f(\cdot)$  es la entrada y $g(\cdot)$ la salida.

### Memoria
Si un sistema LTI no tiene memoria, esto implica que 

$$ h(t-\tau) \neq 0, \text{solo para } t = \tau $$

### Invertibilidad
Un sistema LTI invertible cumple con:

$$ g = f \ast (h \ast h^{-1}) = f $$

lo que implica que 

$$ h \ast h^{-1} = \delta $$

### Estabilidad

Para que un sistema LTI sea estable se necesita que:

$$ \int_{-\infty}^{\infty} |h(\tau)| d\tau < \infty $$

###  Respuesta a entrada cero
$$ f(t) = 0 \hspace{.2cm} \forall t \rightarrow g(t) = 0, \hspace{.2cm}  \forall t $$

Por linealidad:

$$ 0 f(t) \rightarrow 0 g(t) $$

### Causalidad

Un sistema causal implica que 

$$ h(t) = 0, \hspace{1cm}  t < 0 $$


Para un sistema LTI causal, se tiene que

$$ g(t) = \int_{-\infty}^{+t} f(\tau)h(t-\tau) d\tau $$
