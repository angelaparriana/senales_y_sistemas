## El espacio de Schwartz

Recordemos que el [espacio de Schwartz](06-Distribuciones.ipynb#El-espacio-de-Schwartz), $\mathcal{S}(\mathbb{R};\mathbb{C})$ es el espacio de todas las funciones $\varphi: \mathbb{R} \mapsto \mathbb{C}$ que son infinitamente derivables y que son rápidamente decrecientes. 

Es obvio que las funciones de Schwartz son $L^1$. No sólo se trata de funciones integrables sino que además son suaves, es decir, son infinitamente diferenciables, lo que implica que sus Transformadas de Fourier existen y decaen rápido. Pensándolo bien, este espacio es un paraíso para la Transformada de Fourier. 


## Transformada de Fourier en $\mathcal{S}(\mathbb{R};\mathbb{C})$

Es claro que la Transformada de Fourier en $L^1(\mathbb{R};\mathbb{C})$ tiene algunos problemas. En efecto, muchas de las propiedades de la Transformada de Fourier son válidas sólo para un subconjunto del espacio de funciones integrables $L^1$. Por ejemplo, las propiedades de la derivada, inversión, multiplicación y el teorema de Plancherel-Parseval requieren no solo integrabilidad, sino condiciones adicionales para cada caso en particular. Esto no resulta muy conveniente.

Sin embargo, es posible aproximar toda función integrable por funciones que no sólo son integrable, sino que además de comportan perfectamente con respecto a las propiedaes de la Transformada de Fourier en $L^1$. Estas funciones pertenencn el espacio de las funciones de Schwartz $\mathcal{S}(\mathbb{R};\mathbb{C}) \subset L^1(\mathbb{R};\mathbb{C})$. en particular, para cada elemento $f \in  \mathcal{S}(\mathbb{R};\mathbb{C})$,  la multiplicación por $(-i 2 \pi t)^k$ (or $(i 2 \pi u)^k$) para un $k \in \mathbb{N}$ arbitrario, siempre corresponde con la operación de derivar $k$ veces con respecto a $u$ (o $t$) en el espacio recíproco. En otras palabras, es siempre posible derivar respecto al signo de la integral para este tipo de funciones. Adicionalmente, el espacio de Schwartz $\mathcal{S}$ tiene la importante propiedad que es inviariante bajo la Transformada de Fourier. Esto significa que la Transformada de Fourier de cualquier elementos de $\mathcal{S}$ también pertenece a  $\mathcal{S}$ y por lo tanto la fórmula de inversión es siempre aplicable.

## El espacio de Schwartz

Vale la pena recordar la definición formal del espacio de Schwartz. Una función $f:\mathbb{R} \mapsto \mathbb{C}$ se dice que es rápidamente decreciente si decae hacia cero en el infinito más rápido que cualquier potencia inversa. Más precisamente, $\forall k \in \mathbb{N}$, existe $C_k >0$ tal que

$$ \forall \; t \in \mathbb{R}, \;\;\;\;\;\;  |t|^k |f(t)| \leq C_k $$

o, equivalentemente
$$ \sup_{ t \in \mathbb{R}} |t|^k |f(t)| < \infty $$

tal que $t^k f$ está acotada para todo $ k \in \mathbb{N}$.

El espacio de Schwartz $\mathcal{S}(\mathbb{R};\mathbb{C})$ es el espacio de todas las funciones $\varphi: \mathbb{R} \mapsto \mathbb{C}$ que son infinitamente derivables y que son rápidamente decrecientes. 

$$
\boxed{
\mathcal{S}(\mathbb{R};\mathbb{C}) =\left\{ \varphi \in C^{\infty}(\mathbb{R};\mathbb{C}) \; |  \; \; \forall \alpha, k \in \mathbb{N},\;\;\;  \sup_{t \in \mathbb{R}} | t^{\alpha} \frac{d^k \varphi(t)}{dt^k}   | < \infty    \right\}     }
$$

En otras palabras, el espacio de Schwartz consta de todas las funciones $\varphi$ de modo que $\varphi$ y todas sus derivadas existen sobre $\mathbb{R}$ y decaen a cero en $|t| \rightarrow \infty $ más rápido que cualquier potencia inversa de $t$.

- Para $a >0$, una función Gaussiana de la forma $G(t) = e^{-at^2} $, o más generalmente, cualquier producto de un polinomio y una Gaussiana pertenece al espacio de Schwartz.

$$ f(t)= t^k e^{-at^2} \in   \mathcal{S}(\mathbb{R};\mathbb{C}) $$

donde $k \in \mathbb{N}$.

- Si $f \in \mathcal{S}(\mathbb{R};\mathbb{C})$ and $\alpha,k \in \mathbb{N}$, entonces

$$ t^{\alpha} \frac{d^{k} f(t)}{dt^{k}}  \in   \mathcal{S}(\mathbb{R};\mathbb{C})   $$


- Si $f \in   \mathcal{S}(\mathbb{R};\mathbb{C})  $, entonces $f$ es $p$-integrable ($f \in L^p(\mathbb{R};\mathbb{C}) $) para cualquier $p \in [1, \infty]$. En efecto, dado que $f$ está acotada, $f \in L^{\infty}(\mathbb{R};\mathbb{C}) $. Mas aún, existe $C$ tal que $(1+t^2)| f(t) | \leq C$, tal que para $1 \leq p < \infty$

$$  \int_{\mathbb{R}} | f(t) |^p dt \leq C \int_{\mathbb{R}} \frac{dt}{(1+t^2)^p} <\infty \;\;\; \mbox{dado que} \;\;\; 2p > 1$$

y $f \in L^p(\mathbb{R};\mathbb{C})$. En particular, cada elemento de $\mathcal{S}(\mathbb{R};\mathbb{C})$ es integrable ($p=1$) y por lo tanto tiene una Transformada de Fourier bien definida en $L^1$.\\

- Cada función infinitamente derivable $f$ con soporte compacto ($f \in C^{\infty}_{c}(\mathbb{R};\mathbb{C})$) pertenece a $ \mathcal{S}(\mathbb{R};\mathbb{C})$.  Dado que  $C^{\infty}_{c}(\mathbb{R};\mathbb{C})$ es denso en $L^p(\mathbb{R};\mathbb{C})$ para $1\leq p < \infty$, sigue que el espacio Schwartz es denso in $L^p(\mathbb{R};\mathbb{C})$. 

***
## Convergencia en $\mathcal{S}(\mathbb{R};\mathbb{C})$
A pesar que el espacio de Schwartz no constituye un espacio vectorial normado, es posible definir convergencia en  $\mathcal{S}(\mathbb{R};\mathbb{C})$. La secuencia  $(f_n)_{n \geq 0}$ de elementos de $\mathcal{S}(\mathbb{R};\mathbb{C})$ converge a $f \in \mathcal{S}(\mathbb{R};\mathbb{C})$ en $\mathcal{S}(\mathbb{R};\mathbb{C})$ si

\begin{equation}
\forall \alpha, \beta \in \mathbb{N}, \;\;\;\;  \lim_{n \rightarrow \infty} \| x^{\beta} \frac{d^{\alpha}}{dt^{\alpha}} (f_n-f) \|_{L^{\infty}} = 0 
\end{equation}

## La Transformada de Fourier es una bijección lineal  $\mathcal{S}(\mathbb{R};\mathbb{C})  \mapsto \mathcal{S}(\mathbb{R};\mathbb{C})$

El espacio de Schwartz tiene la sorprendente propiedad de ser invariante bajo la Transformada de Fourirer, lo que implica que $\mathcal{F}\left\{ \mathcal{S}(\mathbb{R};\mathbb{C}) \right\} = \mathcal{S}(\mathbb{R};\mathbb{C})$. Más precisamente, la Transformada de Fourier es una bijección lineal de $\mathcal{S}(\mathbb{R};\mathbb{C})$ hacia sí misma.

Es claro que la Transformada de Fourier $\mathcal{F}: \mathcal{S}(\mathbb{R};\mathbb{C}) \rightarrow \mathcal{S}(\mathbb{R};\mathbb{C})$ es injectiva, dado que $\mathcal{S}(\mathbb{R};\mathbb{C}) \subset L^1(\mathbb{R};\mathbb{C})$. Finalmente, dado que en particular $\mathcal{F}\{{f}\} $ es integrable, el teorema de inversión se aplica y 

\begin{equation}
  \mathcal{F}\{\mathcal{F}\{{f}\}\} (u)  = f(-u)  
\end{equation}

y se concluye que $\mathcal{F}: \mathcal{S}(\mathbb{R};\mathbb{C}) \rightarrow \mathcal{S}(\mathbb{R};\mathbb{C}) $ es surjectiva, dado que para toda $f \in \mathcal{S}(\mathbb{R};\mathbb{C})$ es la Transformada de Fourier de un elemento de $\mathcal{S}(\mathbb{R};\mathbb{C})$. 

## Propiedades de la Transformada de Fourier en $\mathcal{S}(\mathbb{R};\mathbb{C})$
Es bastante evidente que las propiedades de desplazamiento, escalamiento, convolución y la fórmula de transferencia, que son válidas para toda función integrable, siguen viendo válidas en el espacio de Schwartz, puesto que $\mathcal{S}(\mathbb{R};\mathbb{C}) \subset L^1(\mathbb{R};\mathbb{C})$. Además, todas las otras propiedades de la Transformada de Fourier discutidas previamente también aplican para $\mathcal{S}(\mathbb{R};\mathbb{C})$, sin ninguna restricción, como veremos a continuación.

***
### Derivada
Si $f \in \mathcal{S}(\mathbb{R};\mathbb{C})$, entonces para cualquier $k \in \mathbb{N}$

\begin{equation}
\boxed{\mathcal{F}\{\frac{d^k  f}{dt^k}\}(u)  = (i 2 \pi u)^k \mathcal{F}\{f\}(u)}
\end{equation}

y

\begin{equation}
\boxed{\frac{d^k \mathcal{F}\{f\}(u)}{d u^k}  = \mathcal{F}\{(-i 2 \pi t)^k f\}(u)}
\end{equation}

Esto es una consecuencia del decaimiento rápido de la Transformada de Fourier y el hecho que $f$ y todas sus derivadas con funciones que decaen rápido.

***
### Inversión
Esto es una consecuencia de la bijección del espacio de Schwartz. Si $f \in  \mathcal{S}(\mathbb{R};\mathbb{C})$,  su Transformada de Fourier también está en $\mathcal{S}(\mathbb{R};\mathbb{C})$ y la fórmula de inversión siempre se aplica:

\begin{equation}
\boxed{f(t)= \int_{\mathbb{R}} e^{+i2\pi u t} \mathcal{F}\{f\}(u) du } \;\;\;\; \forall t \in \mathbb{R}
\end{equation}

o, equivalentemente,
\begin{equation}
  \boxed{\mathcal{F}\{ \mathcal{F}\{f\} \}(u)  = f(-u) } 
\end{equation}

***
### Multiplicación y convolución
Si $f,g \in \mathcal{S}(\mathbb{R};\mathbb{C})$, entonces 

\begin{equation}
\boxed{\mathcal{F}\{ g \ast f\} (u) =  \; \mathcal{F}\{g\}(u) \mathcal{F}\{f\}(u)}
\end{equation}

lo que es verdadero para cada $f,g \in L^1(\mathbb{R};\mathbb{C}) $. Además,

\begin{equation}
\boxed{\mathcal{F}\{fg\}(u) = \mathcal{F}\{f\} \ast \mathcal{F}\{g\}(u)}
\end{equation}

Por ejemplo, para probar la propiedad de multiplicación, la bijección de la Transformada de Fourier sobre $\mathcal{S}(\mathbb{R};\mathbb{C})$ asegura que existen $x,y \in  \mathcal{S}(\mathbb{R};\mathbb{C})$ tal que $f=\mathcal{F}\{x\}$ y $g=\mathcal{F}\{y\}$. Entonces

$$\mathcal{F}\{fg\}(u) = \mathcal{F}\{\mathcal{F}\{x\} \mathcal{F}\{y\}\}(u)  $$

además $ \mathcal{F}\{x\} \mathcal{F}\{y\} = \mathcal{F}\{x \ast y\} $

$$  \mathcal{F}\{fg\} (u) =  \mathcal{F}\{ \mathcal{F}\{ x \ast y \} \} (u) = \{x \ast y\} (-u)  $$

y $\mathcal{F}\{f\} = x(-u)$, $\mathcal{F}\{g\} = y(-u)$ lo que prueba que

$$\mathcal{F}\{fg\}(u) =  \{\mathcal{F}\{f\} \ast \mathcal{F}\{g\}\}(u)$$

***
### Teorema de Plancherel-Parseval

Dado que $\mathcal{S}(\mathbb{R};\mathbb{C}) \subset L^1(\mathbb{R};\mathbb{C})$ y $\mathcal{S}(\mathbb{R};\mathbb{C}) \subset L^2(\mathbb{R};\mathbb{C})$, el teorema de Plancherel-Parseval se aplica en $\mathcal{S}(\mathbb{R};\mathbb{C})$. Para $f,g \in \mathcal{S}(\mathbb{R};\mathbb{C})$:


\begin{equation}
\boxed{(\mathcal{F}\{f\}, \mathcal{F}\{g\}) = ( f,g ); \; \;\;\;\;\; \| \mathcal{F}\{f\} \|_{L^2}^2  = \|  f \|_{L^2}^2   }  
\end{equation}

Vemos que el espacio de Schwartz es el espacio ideal con respecto a las propiedades de la Transformada de Fourier en $L^1$.
